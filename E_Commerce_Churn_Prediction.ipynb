```python
from scipy.stats import pointbiserialr

# Calculate point-biserial correlation coefficient and p-value
corr, p_value = pointbiserialr(df['Complain'], df['Churn'])
print("Point-biserial correlation coefficient:", corr)
print("p-value:", p_value)
```

The point-biserial correlation coefficient measures the strength and direction of the linear relationship between a binary variable (churn) and a continuous variable (customer complaints).
In this case, the coefficient is approximately 0.25, indicating a moderate positive correlation between customer complaints and churn.
The p-value associated with the coefficient is very close to zero (4.28e-81), indicating strong evidence against the null hypothesis that there is no correlation. Thus, we reject the null hypothesis.

Overall, all three statistical tests provide consistent evidence that there is a significant correlation or association between customer complaints and churn in the dataset. The strength of the correlation or association is moderate, suggesting that an increase in customer complaints is associated with a higher likelihood of churn, and vice versa.

```python
from sklearn.metrics import roc_curve, roc_auc_score, confusion_matrix

def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    # Fit the model
    model.fit(X_train, y_train)

    # Make predictions
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    # Calculate probabilities for ROC curve
    y_train_prob = model.predict_proba(X_train)[:, 1]
    y_test_prob = model.predict_proba(X_test)[:, 1]

    # Calculate AUC
    auc_train = roc_auc_score(y_train, y_train_prob)
    auc_test = roc_auc_score(y_test, y_test_prob)

    # Generate ROC curve
    fpr_train, tpr_train, _ = roc_curve(y_train, y_train_prob)
    fpr_test, tpr_test, _ = roc_curve(y_test, y_test_prob)

    # Create confusion matrix
    cm_train = confusion_matrix(y_train, y_train_pred)
    cm_test = confusion_matrix(y_test, y_test_pred)

    # Plot ROC curve
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.plot(fpr_train, tpr_train, label='ROC Curve (AUC = {:.2f})'.format(auc_train))
    plt.plot([0, 1], [0, 1], linestyle='--', color='r', label='Random Guess')
    plt.title('ROC Curve (Train)')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(fpr_test, tpr_test, label='ROC Curve (AUC = {:.2f})'.format(auc_test))
    plt.plot([0, 1], [0, 1], linestyle='--', color='r', label='Random Guess')
    plt.title('ROC Curve (Test)')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()

    plt.tight_layout()
    plt.show()

    # Plot confusion matrix
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    sns.heatmap(cm_train, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title('Confusion Matrix (Train)')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')

    plt.subplot(1, 2, 2)
    sns.heatmap(cm_test, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title('Confusion Matrix (Test)')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')

    plt.tight_layout()
    plt.show()

    # Print AUC and confusion matrix
    print("AUC (Train): {:.2f}".format(auc_train))
    print("AUC (Test): {:.2f}".format(auc_test))
    print("\nConfusion Matrix (Train):\n", cm_train)
    print("\nConfusion Matrix (Test):\n", cm_test)
```
